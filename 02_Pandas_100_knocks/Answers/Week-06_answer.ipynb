# マージと連結(59 - 65)

```python
# 【59】
# df2にdf3を左結合し、df2に格納

df3
```

```python
df2 = pd.read_csv('./data/data1.csv')
df2 = pd.merge(df2,df3,on='name',how='left')
df2
```

・左結合では、df2に存在するレコードにdf3のレコードを結合する  
・on=''で結合キーを指定  
・how=''で結合方法を指定

```python
# 【60】
# df2にdf3を右結合し、df2に格納
df2 = pd.read_csv('./data/data1.csv')
df2 = pd.merge(df2,df3,on='name',how='right')
df2
```

・右結合では、df3に存在するレコードにdf2のレコードを結合する  
・on=''で結合キーを指定  
・how=''で結合方法を指定

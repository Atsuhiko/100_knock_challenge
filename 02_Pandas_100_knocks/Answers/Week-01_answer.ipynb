## 参照渡し (shallow copy ) と値渡し (deep copy)

```python
df_1 = pd.DataFrame(['a', 'b', 'c']).T #行と列を入れ替える
df_1
```

```python
# shallow copy（参照渡し）
df_2 = df_1
df_2[1] = "apple"
df_2
```

```python
# shallow copy をするとコピー先を変更しても元の変更されてしまう
df_1
```

```python
# deep copy (値渡し)
df_1 = pd.DataFrame(['a', 'b', 'c']).T #行と列を入れ替える
df_2 = df_1.copy()
```

```python
df_2[0] = "orange"
df_2
```

```python
# 元データは変化しない
df_1
```

```python
# 【8】
# ① dfのpclassの列のデータ型をdtypeで確認
# ② 数値型から文字型に変換し、データ型をdtypeで確認

print(df['pclass'].dtype)
df['pclass'] = df['pclass'].astype(str)
print(df['pclass'].dtype)
```

・データ型を変更する場合は astype を使用  
・問題40の列同士の結合では、数値列と文字列を結合しますが、データ型が異なると結合が上手くいきません。  
両方の列のデータ型が同じになるようにastypeを使用してデータ型の変換をします。

```python
# ライブラリーの読み込み
import pandas as pd
import glob

# DataFrameの準備
df = pd.read_csv('./data/titanic3.csv')
df2 = pd.read_csv('./data/data1.csv')
df3 = pd.read_csv('./data/data1_2.csv')
df4 = pd.read_csv('./data/data1_3.csv')
df5 = pd.read_csv('./data/data2.csv',encoding='cp932')
```

# マージと連結(59 - 65)

```python
df2.head()
```

```python
df3.head()
```

```python
# 【61】
# df2にdf3を内部結合し、df2に格納

df2 = pd.merge(df2,df3,on='name',how='inner')
df2
```

・内部結合では、df2とdf3の共通のキーのみで結合する  
・on=''で結合キーを指定  
・how=''で結合方法を指定  

```python
# 【62】
# df2にdf3を外部結合し、df2に格納

df2 = pd.read_csv('./data/data1.csv')
df2 = pd.merge(df2,df3,on='name',how='outer')
df2
```

・外部結合では、df2とdf3の両方に存在するレコードが残るように結合する  
・on=''で結合キーを指定  
・how=''で結合方法を指定

```python
df4.head()
```

```python
# 【63】
# df2とdf4を列方向に連結し、df2に格納

df2 = pd.read_csv('./data/data1.csv')
df2 = pd.concat([df2,df4],axis=1)
df2
```

・複数のデータフレームを連結するときはpd.concatを使用  
・axis=0で行方向、axis=1で列方向に連結  
・pd.concat([df2,df4],axis=1)でdf2とdf4を列方向に連結

```python
# 【64】
# df2とdf4を列方向に連結後、重複している
# name列の片方を削除し、df2に格納

df2 = pd.read_csv('./data/data1.csv')
df2 = pd.concat([df2,df4],axis=1)
df2 = df2.loc[:,~df2.columns.duplicated()]
df2
```

df2.loc[:,~df2.columns.duplicated()]により重複した列を消去

```python
# 【65】
# df2とdf2を行方向に連結し、df2に格納

df2 = pd.read_csv('./data/data1.csv')
df2 = pd.concat([df2,df2],axis=0)
df2
```

・複数のデータフレームを連結するときはpd.concatを使用  
・axis=0で行方向、axis=1で列方向に連結  
・pd.concat([df2,df2],axis=0)でdf2とdf2を行方向に連結

```python
# 【68】
# ①df2の生徒ごとの合計点（行方向の合計）
# ②df2の科目ごとの点数の総和（列方向の合計）

df2 = pd.read_csv('./data/data1.csv')
df2 = df2.drop(['class'],axis=1)
print(df2.sum(axis=1)) #行方向の合計
print(df2.sum()) #列方向の合計

```

・合計値の確認はsum()を使用  
・引数空欄の場合、デフォルトは列方向の合計  
・引数にaxis=1を指定すると行方向の合計

```python
# 【23】
# ilocを使ってdfのage列を5行目まで表示

df.iloc[:5,4]
```

```python
# 【24】
# dfのname,age,sexの列のみ抽出しdf_copyに格納
# その後outputフォルダにcsvファイルで出力
# 出力ファイル名はsmaple.csv

df_copy = df[['name','age','sex']]
df_copy.to_csv('./data/sample.csv')
```

```python
# きちんと保存できているか確認
df_copy2 = pd.read_csv('./data/sample.csv')
df_copy2.head()
```

```python
# 【25】
# dfのage列の値が30以上のデータのみ抽出

df[df['age'] >= 30]
```

```python
# df[] の [] 中に条件（この場合 df['age'] >= 30）を書けるのは超便利です！
```

```python
# 【26】
# dfのsex列がfemaleのデータのみ抽出

df[df['sex'] == 'female']
```

```python
# 【27】
# dfのsex列がfemaleでかつageが40以上のデータのみ抽出

df[(df['sex'] == 'female' ) & (df['age'] >= 40)]
```

```python
# df[] の中に、条件の組み合わせも入れることができませす。
```

```python
# 【28】
# queryを用いてdfのsex列がfemaleでかつageが40以上のデータのみ抽出

df.query('sex == "female" & age >= 40 ')
```

```python
# 同じことは、query() を使ってもできます。
# query() には正規表現も使えるので、相当凝った抽出もできます。
# このあたりのコードは、「データサイエンス１００本ノック」を見てください。

# 例えば、以下の問題は
# P-016: 店舗データフレーム（df_store）から、電話番号（tel_no）が3桁-3桁-4桁のデータを全項目表示せよ。

# このように解けます。
# df_store.query("tel_no.str.contains('[0-9]{3}-[0-9]{3}-[0-9]{4}', regex=True)", engine='python')
```

```python
# 【29】
# dfのname列に文字列「Mrs」が含まれるデータを表示

df.query('name.str.contains("Mrs")', engine='python')
```

```python
# [Tips]
# 特定の文字列を含むデータを抽出したいときは
# str.contains()を使用
# engine='python'を指定しないとエラーが出る
```

```python
# query() で文字列の検索もできるのは便利です！
```

```python
# 【30】
# dfの中で文字型の列のみを表示

df.select_dtypes(include='object')
```

```python
# [Tips]
# ・特定のデータ型の列を抽出したい時は
# 　select_dtypesを使用する。
# ・今回は文字列を抽出したいのでinclude='object'を
# 　指定
# ・exclude='object'をすれば数値型の列を抽出可能

# ex)
# df.select_dtypes(exclude='object')
```

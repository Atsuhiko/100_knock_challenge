```python
# ライブラリーの読み込み
import pandas as pd
import glob

# DataFrameの準備
df = pd.read_csv('./data/titanic3.csv')
df2 = pd.read_csv('./data/data1.csv')
df3 = pd.read_csv('./data/data1_2.csv')
df4 = pd.read_csv('./data/data1_3.csv')
df5 = pd.read_csv('./data/data2.csv',encoding='cp932')
```

# データ加工 (33 - 58)

```python
# 【51】
# dfの行をシャッフルし、インデックスを振り直して表示

df.sample(frac=1)


```

```python
# 【52】
# ①df2の重複行数をカウント
# ②df2の重複行を削除し、df2を表示

print(df2.duplicated().value_counts())
df2 = df2.drop_duplicates()
df2


```

```python
# 【53】
# dfのnameの列をすべて大文字に変換し表示


df['name'].str.upper()

```

```python
# 【54】
# dfのnameの列をすべて小文字に変換し表示

df['name'].str.lower()

```

```python
# 【55】
# dfのsex列に含まれる「female」という単語を
# 「Python」に置換。その後、1行目の
# 「female」が「Python」に置き換わったことを確認

df['sex'] = df['sex'].replace('female','Python')
df.head()


```

```python
# 【56】
# dfのname列1行目の「Allen, Miss. Elisabeth Walton」の
# 「Elisabeth」を消去(import reをインポート)

・完全一致で文字列を消去するときはreplaceを使用
.部分一致はre.sub

import re
df['name'][0] = re.sub('Elisabeth','',df['name'][0])
df['name'][0]


```

```python
# 【57】
# df5の都道府県列と市区町村列を空白がないように
# 「_」で結合(新規列名は「test2」)し、先頭5行を表示
# ※df5の「test」列は通常通り結合した場合の結果


df5['test2'] = df5['都道府県'].str.rstrip() +'_'+ df5['市区町村']
df5.head()
```

```python
# 【58】
# df2の行と列を入れ替えて表示


df2 = df2.transpose()
df2


```

# マージと連結(59 - 65)

```python
# 【59】
# df2にdf3を左結合し、df2に格納

df2 = df2.transpose()

df2 = pd.merge(df2,df3,on='name',how='left')
df2



```

```python
# 【60】
# df2にdf3を右結合し、df2に格納


df2 = pd.merge(df2,df3,on='name',how='right')
df2

```

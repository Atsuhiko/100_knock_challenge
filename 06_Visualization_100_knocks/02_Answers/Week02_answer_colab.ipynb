## Q10. dfより以下のような散布図（scatter plot）を描きなさい。

```python
df.plot.scatter(x='a1', y='a3', alpha=0.5)
```

## オマケコード１. dfより六角形ビニング図（hexbin plot）を描きなさい。

```python
df.plot.hexbin(x='a1', y='a3',
               gridsize=15, sharex=False)
```

## オマケコード２. 円グラフ（pie plot）

```python
df_pie = pd.DataFrame([[1, 50], [2, 20], [3, 30]],
                      index=['a', 'b', 'c'], columns=['ONE', 'TWO'])
df_pie.plot.pie(subplots=True)
```

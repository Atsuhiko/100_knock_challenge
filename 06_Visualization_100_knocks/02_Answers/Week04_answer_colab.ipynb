## Q18. 以下のように三角関数を描画する関数を作成して、Sin, Cos, Tan 曲線を描きなさい。

image.png

```python
# Numpy と Matplotlib を用いた三角関数を描画

def Trigonometric_Plot(i):
    x = np.linspace(-i*np.pi, i*np.pi)
    plt.plot(x, np.cos(x), color='r', ls='-', label='cos')
    plt.plot(x, np.sin(x), color='b', marker='p', ls=':', label='sin')
    plt.plot(x, np.tan(x), color='c', marker='s', ls='-', label='tan')

    plt.xlim(-np.pi, np.pi)
    plt.ylim(-1.5, 1.5)

    plt.axhline(0, ls='-', c='b', lw=0.5)
    plt.axvline(0, ls='-', c='b', lw=0.5)

    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Trigonometric functions')

    plt.show()
```

```python
Trigonometric_Plot(3)
```

## Q19. ディープラーニングの学習曲線を描く関数を作成し描画しなさい。

image.png

image.png

```python
# Tensorflow/Keras の historyクラスを定義
class history:
    def __init__(acc, val_acc, loss, val_loss):
        self.acc = acc
        self.val_acc = val_acc
        self.loss = loss
        self.val_loss = val_loss

# historyクラスにメンバーを入力
his = history
his.acc = [0.50376884, 0.75753769, 0.85050251, 0.91708543, 0.93592965]
his.val_acc = [0.665, 0.875, 0.86, 0.955, 0.975]
his.loss = [1.0869408207922127, 0.612384136298194, 0.4129973124319585, 0.2651988666261261, 0.20133653537711904]
his.val_loss = [0.7168122148513794, 0.4103385591506958, 0.3806729394197464, 0.15852605164051056, 0.11399855181574821]
```

```python
# 関数 Learning_Process_Plot() を定義 --> 引数は historyクラス
def Learning_Process_Plot(history):
    acc = history.acc
    val_acc = history.val_acc
    loss = history.loss
    val_loss = history.val_loss

    epochs = range(len(acc))

    plt.plot(epochs, acc, label='Training acc', ls="-", marker="o")
    plt.plot(epochs, val_acc, label='Validation acc', ls="-", marker="x")
    plt.title('Training and validation accuracy')
    plt.ylabel("accuracy")
    plt.xlabel("epoch")
    plt.legend(loc="best")
    plt.savefig('acc_graph.png') # acc, val_acc グラフ保存

    plt.figure()

    plt.plot(epochs, loss, label='Training loss', ls="-", marker="o")
    plt.plot(epochs, val_loss, label='Validation loss', ls="-", marker="x")
    plt.title('Training and validation loss')
    plt.ylabel("loss")
    plt.xlabel("epoch")
    plt.legend(loc="best")
    plt.savefig('loss_graph.png') # loss, val_loss グラフ保存

    plt.show()
```

```python
# 関数Learning_Process_Plot()
Learning_Process_Plot(his)
```

## Q20. ガウス混合モデルによるクラスタリングの結果を以下のように散布図と確率密分布を表示しなさい。

```python
# Pomegranateによる混合ガウスモデルの実装
# https://github.com/jmschrei/pomegranate/tree/master/tutorials

from pomegranate import *
np.random.seed(0)

# ３つのクラスターを想定した模擬データの作成
X = np.concatenate([np.random.normal((7, 2), 1, size=(100, 2)),
                       np.random.normal((2, 3), 1, size=(150, 2)),
                       np.random.normal((7, 7), 1, size=(100, 2))])
```

```python
# データの可視化
plt.figure(figsize=(8, 6))
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
# グリッドの作成
x = np.arange(-1, 10.1, .1)
y = np.arange(-1, 10.1, .1)

xx, yy = np.meshgrid(x, y)
x_ = np.array(list(zip(xx.flatten(), yy.flatten())))

# ３成分の混合ガウスモデル作成
n_components = 3
model = GeneralMixtureModel.from_samples(MultivariateGaussianDistribution, n_components, X)

# 確率分
p = model.probability(x_).reshape(len(x), len(y))
p
```

```python
# クラスタリング
pred = model.predict(X)

# プロット用のデータフレーム作成
df_pome = pd.DataFrame(X, columns = ["x", "y"])
df_pome["cluster"] = pred
df_pome.head()
```

```python
# 確率分布の可視化
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, p, cmap='Blues', alpha=0.8)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
# クラスター可視化
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(1,1,1) 
targets = [0,1,2]
colors = ['r', 'b', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = df_pome["cluster"] == target
    ax.scatter(df_pome.loc[indicesToKeep, 'x'],
               df_pome.loc[indicesToKeep, 'y'],
               c = color,
               s = 50)
ax.legend(targets)
plt.show()
```

```python
# クラスターと確率分布の可視化
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(1,1,1) 
targets = [0,1,2]
colors = ['r', 'b', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = df_pome["cluster"] == target
    ax.scatter(df_pome.loc[indicesToKeep, 'x'],
               df_pome.loc[indicesToKeep, 'y'],
               c = color,
               s = 50)
ax.legend(targets)
ax.contourf(xx, yy, p, cmap='Blues', alpha=0.7)
plt.show()
```

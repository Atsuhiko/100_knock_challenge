## Q１～Q2：折れ線グラフ  
データセット：Pulsar Stars（簡略版）  
https://www.kaggle.com/datasets/colearninglounge/predicting-pulsar-starintermediate

```python
# CSVファイルの読み込み
# Pulsar stars のデータセット
df = pd.read_csv("data/pulsar_stars_s.csv") # df = data frame の略
df.head()
```

## Q1. dfから３本の折れ線グラフを一つの領域に描きなさい

image.png

```python
# 列と行を指定してプロットする
df.plot() # グラフ描画
```

```python
df["Mean_IP"].plot() 
```

```python
df["Mean_DM-SNR"].plot() 
```

```python
df[["Mean_IP","SD_IP"]].plot() 
```

## Q2. dfより各列の折れ線グラフをそれぞれ別のグラフに分けてに描画しなさい。

image.png

```python
df.plot(subplots=True)
```

```python
# 軸、y軸の範囲が共通化
df.plot(subplots=True, sharex=True, sharey=True) 
```

## Q3. dfより３本の折れ線グラフを表示方法を変えて描画しなさい。

image.png

```python
# 折れ線グラフの線の種類を変える
df.plot.line(style=['r--', 'b.-', 'g+'])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Iris dataset

```python
# CSVファイルの読み込み
df = pd.read_csv("./data/iris.csv") # df = dataframe の略
df.head()
```

## Q21. Seaborn を用いて以下の様な pairplot（各特徴量の分布と特徴量間の散布図）を描画せよ。

image.png

```python
sns_plot = sns.pairplot(df, hue ="label", height=2.5)
```

## Q22. Seabornを用いて以下の様な特徴量間の相関係数のヒートマップを描画せよ。

image.png

```python
#ヒートマップの表示,annot=Trueにすると数値出せる
fig, ax = plt.subplots(figsize=(6, 6))
sns_heatmap=sns.heatmap(df.corr(), cmap=sns.color_palette("coolwarm",10), annot=True, fmt=".2f", vmin=-1, vmax=1,ax=ax, square=True)
plt.savefig("heatmap.png", bbox_inches="tight", dpi=120)
#sns_heatmap.savefig("heatplot.png")
```

## Q23. Seabornを用いて以下のような、特徴量a1とa4の lineplot を描画せよ。

image.png

```python
sns.lineplot(x="a1", y="a4", data=df)
```

## 24. Seabornを用いて以下の様な特徴量a1とa4の barplot を描画せよ。

image.png

```python
sns.barplot(x="a1", y="a4", data=df)
```

## 25. Seabornを用いて以下の様な特徴量a1とa4の Implot を描画せよ。

image.png

```python
sns.lmplot(x="a1", y="a4", data=df, hue="label")
```

> P-039: レシート明細データフレーム（df_receipt）から売上日数の多い顧客の上位20件と、売上金額合計の多い顧客の上位20件を抽出し、完全外部結合せよ。ただし、非会員（顧客IDが'Z'から始まるもの）は除外すること。

```python
# いままでの定番のパターンです。
df_date = df_receipt.groupby('customer_id').amount.sum().reset_index()
df_date2 = df_date.query('not customer_id.str.startswith("Z")', engine='python')
df_date3 = df_date2.sort_values('amount', ascending=False).head(20)
df_date3
```

```python
df_value = df_receipt.groupby('customer_id').sales_ymd.count().reset_index()
df_value2 = df_value.query('not customer_id.str.startswith("Z")', engine='python')
df_value3= df_value2.sort_values('sales_ymd', ascending=False).head(20)
df_value3
```

```python
pd.merge(df_date3, df_value3, how='outer', on='customer_id')
```

```python
# 公式解答と違う！
# これが公式解答
df_sum = df_receipt.groupby('customer_id').amount.sum().reset_index()
df_sum = df_sum.query('not customer_id.str.startswith("Z")', engine='python')
df_sum = df_sum.sort_values('amount', ascending=False).head(20)

df_cnt = df_receipt[~df_receipt.duplicated(subset=['customer_id', 'sales_ymd'])] # これはなんだ？
df_cnt = df_cnt.query('not customer_id.str.startswith("Z")', engine='python')
df_cnt = df_cnt.groupby('customer_id').sales_ymd.count().reset_index()
df_cnt = df_cnt.sort_values('sales_ymd', ascending=False).head(20)

pd.merge(df_sum, df_cnt, how='outer', on='customer_id')

# Su Ma さん解読してください！
```
